```python
import pandas as pd
data = pd.read_csv("./nyc_squirrels.csv")
print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))
data.head()
```

```python
import numpy as np
print('Number of rows before discarding duplicates = %d' % (data.shape[0]))
data = data.drop_duplicates(subset=['unique_squirrel_id'])
print('Number of rows after discarding duplicates = %d' % (data.shape[0]))
data.head()
```

```python
print('Number of instances = %d' % data.shape[0])
print('Number of attributes = %d' % data.shape[1])

print('Number of missing values:')
for col in data.columns:
    print('\t%s: %d' % (col, data[col].isna().sum()))
```

```python
columns_to_drop = ['specific_location', 'highlight_fur_color', 'specific_location','other_activities','other_interactions','zip_codes','color_notes']
data = data.drop(columns=columns_to_drop)
```

```python
print('Number of rows in original data = %d' % (data.shape[0]))

data = data.dropna()

print('Number of rows after discarding missing values = %d' % (data.shape[0]))
```
